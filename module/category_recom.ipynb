```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
p_df = pd.read_csv('./data./게시글.csv')
r_df = pd.read_csv('./data./레시피.csv')
i_df = pd.read_csv('./data./재료.csv')
```

```python
n_df = pd.merge(left = p_df, right = r_df, how = "outer", on = "게시글 id" )
a_df = pd.merge(left = n_df, right = i_df, how = "outer", on = "게시글 id" )
a_df.head(10)
```

```python
p_df.head()
```

```python
ohe = OneHotEncoder(sparse=False)
category_cols = ['종류별 카테고리', '상황별 카테고리', '재료별 카테고리']
encoded_data = ohe.fit_transform(p_df[category_cols])

encoded_df = pd.DataFrame(encoded_data, columns=ohe.get_feature_names_out(category_cols))
p_dff = pd.concat([p_df.drop(columns=category_cols), encoded_df], axis=1)

p_dff
```

```python
okt = Okt()
for i in range(200):
    p_dff.loc[i, '게시글 제목'] = okt(p_dff.loc[i, '게시글 제목'])

p_dff.head()
```

```python
# 게시글 제목 전처리
import re
def clean_chars(text):
    pattern = r"[^ㄱ-ㅎㅏ-ㅣ가-힣]"
    return re.sub(pattern, '', text)

for i in range(200):
    p_dff.loc[i, '게시글 제목'] = clean_chars(p_dff.loc[i, '게시글 제목'])

p_dff.head()
```

```python
tfidf = TfidfVectorizer()
```

```python
titles_list = p_dff['게시글 제목'].tolist()
```

```python
tfidf_matrix = tfidf.fit_transform(titles_list)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
tfidf_df
```
