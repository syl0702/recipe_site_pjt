```python
tfidf = TfidfVectorizer()
all_item_tfidf_vector = tfidf.fit_transform(data).toarray()
```

```python
cnt_vect = CountVectorizer()
cnt_vectors = cnt_vect.fit_transform(data).toarray()
cnt_vectors
```

```python
base_data = cnt_vectors * 23 / np.exp(tfidf.idf_)
base_data
```

```python
sample = {
    '제육볶음': ['한식', '일상', '돼지고기'],
    '된장찌개': ['한식', '원팬스피디', '콩견과류'],
    '토마토파스타': ['양식', '원팬스피디', '면류'],
}


sample_data = sum(sample.values(), [])
sample_data
```

```python
selected_category_vector = np.zeros(cnt_vectors.shape[1])
selected_category_vector
```

```python
for term in sample_data:
    idx = cnt_vect.vocabulary_[term]
    selected_category_vector[idx] = (23 / np.exp(tfidf.idf_))[idx]
selected_category_vector
```

```python
cnt_vect.vocabulary_[term]
```

```python
sims = cosine_similarity(cnt_vectors, selected_category_vector.reshape(1, -1))
sims
```
